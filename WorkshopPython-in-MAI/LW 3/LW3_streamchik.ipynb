Ниже функция заполнения нулевых строк чем-то осозанным. Что-то осознанное - это параметры этой функции. Например в пустые строки с жанром вписываем часто встречающийся жанр

```python
def fill_nan_by_ctgr(df, column, grouped_column, method, dictionary=False):
    gb = df.groupby(grouped_column).apply(method)[column]
    if dictionary == True:
        gb = {ind: gb[ind][0] for ind in gb.index.levels[0]}
    
    col_nan = df[df[column].isna()]
    
    for ind in tqdm(col_nan.index, desc=f'Fillna column: {column}'):
        df[column][ind] = gb[df[grouped_column][ind]]
```

```python
data_streamchik.head(5)
```

```python
fill_nan_by_ctgr(data_streamchik, 'year_of_release', 'platform', pd.DataFrame.median)

data_streamchik['genre'] = data_streamchik['genre'].fillna(data_streamchik['genre'].mode()[0])

fill_nan_by_ctgr(data_streamchik, 'critic_score', 'genre', pd.DataFrame.mean)
fill_nan_by_ctgr(data_streamchik, 'rating', 'genre', pd.DataFrame.mode, dictionary=True)
```
